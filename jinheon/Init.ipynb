```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
df_horse = pd.read_csv('../data/horse.csv')
```

```python
df_horse.head(10)
```

```python
df_horse.shape
```

```python
dict_col_null_count = {}
```

```python
for column in df_horse.columns:
    dict_col_null_count[column] = len(df_horse) - df_horse[column].isnull().sum()
```

```python
pd.DataFrame(dict_col_null_count, index = ['null_count']).sort_values(by = ['null_count'], axis = 1)
```

```python
pd.DataFrame(dict_col_null_count, index=['null_count']).sort_values(by = ['null_count'], axis = 1).plot(kind='bar', figsize = (20, 10))
```

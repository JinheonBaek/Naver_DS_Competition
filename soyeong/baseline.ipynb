## Import....

```python
import numpy as np
import pandas as pd
```

```python
df_horse = pd.read_csv('../data/horse.csv')
```

## Data

```python
df_horse.head()
```

## Get median & mode
* String: mode
* Numeric: median

```python
df_horse.columns
```

```python
dict_column_median_mode = {}

for column in df_horse.columns:
    df_horse_drop_nan_row = df_horse[column].dropna()
    
    if isinstance(df_horse_drop_nan_row.values[0], str):
        dict_column_median_mode[column] = df_horse_drop_nan_row.mode()[0]
    else:
        dict_column_median_mode[column] = df_horse_drop_nan_row.median()
```

```python
dict_column_median_mode
```

## Fill Nan with median & mode
* String: mode
* Numeric: median

```python
for column in df_horse.columns:
    df_horse[column].fillna(dict_column_median_mode[column], inplace = True)
```

```python
df_horse.head(20)
```

## Check if everything is not nan

```python
dict_col_null_count = {}
```

```python
for column in df_horse.columns:
    dict_col_null_count[column] = len(df_horse) - df_horse[column].isnull().sum()
```

```python
dict_col_null_count
```

## String to get_dummies

```python
df_horse = df_horse.reindex(np.random.permutation(df_horse.index))
```

```python
df_horse.head(20)
```

```python
df_horse_one_hot = pd.get_dummies(df_horse.drop(columns = ['outcome']))
```

```python
df_horse_one_hot.head()
```

```python
X = df_horse_one_hot
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
le.fit(df_horse['outcome'])
```

* Lived: 1
* euthanized: 0
* Died: -1

```python
y = le.transform(df_horse['outcome']) - 1
```

## Run ML

```python
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
names = ["OneR", "Nearest Neighbors", "Linear SVM", "RBF SVM",
         "Decision Tree", "Random Forest", "Neural Net", "AdaBoost", "Naive Bayes"]
```

```python
classifiers = [
    DecisionTreeClassifier(max_depth=1),
    KNeighborsClassifier(5),
    SVC(kernel="linear", C=0.025),
    SVC(kernel="rbf", C=0.025),
    DecisionTreeClassifier(max_depth=10),
    RandomForestClassifier(max_depth=10, n_estimators=300),
    MLPClassifier(alpha=1, hidden_layer_sizes = (100, 50)),
    AdaBoostClassifier(n_estimators = 300),
    GaussianNB()]
```

## CV

```python
from sklearn.model_selection import cross_validate
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import precision_recall_fscore_support
```

```python
evaluations = []

for name, clf in zip(names, classifiers):
    y_pred = cross_val_predict(clf, X, y, cv = 10)
    print(len(y_pred))
    
    precision_recall_f1 = precision_recall_fscore_support(y, y_pred, average='weighted')
    scores = cross_validate(clf, X, y, cv = 10)
    print(scores['test_score'])
    
    evaluations.append([name, scores['test_score'].mean(), precision_recall_f1[:3]])
```

## Result

```python
evaluations.sort(key = lambda evaluations: evaluations[1])
evaluations.reverse()
```

```python
for evaluation in evaluations:
    print("-----" * 5)
    print(evaluation[0])
    print("Accuracy: ", evaluation[1])
    if len(evaluation) > 2: print("Precision, Recall, F1", evaluation[2])
    print("-----" * 5, end = "\n\n")
```
